```python
#Import the Ontime masterlist and the Population estimates per LGA
import pandas as pd

Ontime = pd.read_csv('C:\\Users\\Ayomide Joe\\Downloads\\OnTIME.csv')
LG_Pop = pd.read_csv('C:\\Users\\Ayomide Joe\\Downloads\\LG_Pop.csv')
```

```python
Ontime.head()
```

```python
# Group by 'LGA' and sum the values of 'cemoc_verified'
cemoc_aggregated = Ontime[Ontime['cemoc_verified'] == 1].groupby('lga').size().reset_index(name='cemoc_verified_1_count')

# Display the new dataframe
cemoc_aggregated
```

```python
#For LGA counts with Public owners
# Get a list of all unique LGAs
all_lgas = Ontime['lga'].unique()

# Filter rows where both 'cemoc_verified' and 'owner' are 1
filtered_df_owner = Ontime[(Ontime['cemoc_verified'] == 1) & (Ontime['owner'] == 1)]

# Group by 'LGA' and count occurrences, reindex to include LGAs with 0 counts
cemoc_owner_aggregated = filtered_df_owner.groupby('lga').size().reindex(all_lgas, fill_value=0).reset_index(name='cemoc_verified_and_owner_1_count')

# Display the new dataframe
cemoc_owner_aggregated
```

```python
# Get a list of all unique LGAs
all_lgas = Ontime['lga'].unique()

# Filter rows where both 'cemoc_verified' and 'operating_times' are 1
filtered_df = Ontime[(Ontime['cemoc_verified'] == 1) & (Ontime['operating_times'] == 1)]

# Group by 'LGA' and count occurrences, reindex to include LGAs with 0 counts
cemoc_operating_aggregated = filtered_df.groupby('lga').size().reindex(all_lgas, fill_value=0).reset_index(name='cemoc_verified_and_operating_times_1_count')

# Display the new dataframe
cemoc_operating_aggregated
```

```python
# Get a list of all unique LGAs
all_lgas = Ontime['lga'].unique()

# Filter rows where 'cemoc_verified' is 1 and 'specific_owner' is 4 or 5 (for Non-Profits)
filtered_df_specific_owner = Ontime[(Ontime['cemoc_verified'] == 1) & (Ontime['specific_owner'].isin([4, 5]))]

# Group by 'LGA' and count occurrences, reindex to include LGAs with 0 counts
cemoc_NP_owner_aggregated = filtered_df_specific_owner.groupby('lga').size().reindex(all_lgas, fill_value=0).reset_index(name='cemoc_verified_and_specific_owner_4_or_5_count')

# Display the new dataframe
cemoc_NP_owner_aggregated
```

```python
#Areas with Secondary and Tertiary
# Get a list of all unique LGAs
all_lgas = Ontime['lga'].unique()

# Filter rows where 'cemoc_verified' is 1 and 'facility_level' is 1 & 2 for secondary and tertiary
filtered_df_facility_level = Ontime[(Ontime['cemoc_verified'] == 1) & (Ontime['facility_level'].isin([1, 2]))]

# Group by 'LGA' and count occurrences
cemoc_facility_level_aggregated = filtered_df_facility_level.groupby('lga').size().reindex(all_lgas, fill_value=0).reset_index(name='cemoc_verified_and_facility_level_2_count')

# Display the new dataframe
cemoc_facility_level_aggregated
```

```python
# Merge cemoc_aggregated with cemoc_operating_aggregated
merged_df = pd.merge(cemoc_aggregated, cemoc_operating_aggregated, on='lga', how='left')

# Merge the result with cemoc_owner_aggregated
merged_df = pd.merge(merged_df, cemoc_owner_aggregated, on='lga', how='left')

# Merge the result with cemoc_NP_owner_aggregated
merged_df = pd.merge(merged_df, cemoc_NP_owner_aggregated, on='lga', how='left')

# Finally, merge the result with cemoc_facility_level_aggregated
merged_df = pd.merge(merged_df, cemoc_facility_level_aggregated, on='lga', how='left')

# Display the final merged dataframe
merged_df
```

```python
#Merge all with LG_Pop

# Clean 'lga' in merged_df: Convert to lowercase and strip extra characters
merged_df['lga_cleaned'] = merged_df['lga'].str.lower().str.strip().str.replace('[^a-z0-9 ]', '', regex=True)

# Clean 'STATE/LGA' in LG_Pop: Convert to lowercase and strip extra characters
LG_Pop['STATE/LGA_cleaned'] = LG_Pop['STATE/LGA'].str.lower().str.strip().str.replace('[^a-z0-9 ]', '', regex=True)

# Merge on the cleaned columns
merged_pop = pd.merge(merged_df, LG_Pop, left_on='lga_cleaned', right_on='STATE/LGA_cleaned', how='left')

# Drop the cleaned columns if you don't need them anymore
merged_pop.drop(columns=['lga_cleaned', 'STATE/LGA_cleaned'], inplace=True)

# Display the merged dataframe
merged_pop
```
